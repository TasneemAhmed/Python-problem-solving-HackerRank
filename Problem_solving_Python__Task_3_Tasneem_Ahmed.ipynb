```python
class person:
  height = 160

  def __init__(self, name, age):
    self.name = name
    self.age = age

    self.price = 50000
    self.__maxPrice = 70000  #__this make the attribute private, can't change it's value 

 # Change the private attribute only through this method
  def setMaxPrice(self, price):
    self.__maxPrice = price

  def sell(self):
    print(f'{self.price} -> {self.__maxPrice}')

  def work(self, salary):
    print(f'{self.name} -> {salary}')
```

```python
p1 = person('Tasneem', 24)
p1.work(12000)
```

```python
class person2(person):
  def job():
    print("I am Big Data Engineer")
```

```python
p2 = person2('Mostafa', 30)

p2.work(30000)
p2.job
```

```python
# This example for Polymorphism
# have the same name function fly() in diffent clases with diff behaviour
class parrot():
  def fly(self):
    print("Parrot can fly")

class penguin():
  def fly(self):
    print("Parrot can't fly")

# send object to class and accesse fly() function
def test_fly(object):
  object.fly()
```

```python
pa = parrot()
pe = penguin()

test_fly(pa)
test_fly(pe)
```

```python
# Overriding method: happens between super & sub class
  # Have the same signature(name, parameters) but different behaviour

class car():
  def __init__(self, name, brand):
    self.car_name = name
    self.car_brand = brand

class car_copy(car):
  def __init__(self):
    car.__init__(self,'Tyota Crolla', 'Tyota') # Override __init__ method, assign default values to parameters
```

```python
c1 = car_copy()

c1.car_brand
c1.car_name
```

```python
isinstance(c1, car)
```

```python
isinstance(c1, car_copy)
```

```python
issubclass(car_copy, car)
```

```python
issubclass(car, car_copy)
```

```python
# Exercise 6 : Class inheritance & apply Override
class Vehicle:
  def __init__(self, name, mileage, capacity):
    self.vehicleName = name 
    self.mileAge = mileage
    self.vehicleCapacity = capacity

  # Create fare() method
  def fare(self):
    return self.vehicleCapacity * 100

class Bus(Vehicle):
  # need to override(change behaviour/return of function) the fare() method which in superClass(Vehicle)
  # Method overrding allows a child class to provide a specific implementation of a method that is already provided by one of its parent class.
  def fare(self):
    return (self.vehicleCapacity * 100) + (self.vehicleCapacity * 100 * .1)

bus = Bus("School Volvo", 12, 50)
print("Total Bus fare is {}".format(bus.fare()))
```
